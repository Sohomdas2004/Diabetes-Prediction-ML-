```python
import pandas as pd
import numpy as np
```

```python
#Loading the data set and showing the first rows of the dataset
df=pd.read_csv("/diabetes.csv")
df.head()
```

```python
#Describes every column of the dataset
df.describe()
```

```python
#checking the infrormation(data type) of every column
df.info()
```

```python
#Checking the missing values
df.isnull().sum()
```

```python
#Fill Missing Values
cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']

df[cols] = df[cols].replace(0, np.nan)

for col in cols:
    df[col] = df[col].fillna(df[col].median())

```

```python
df.head()
```

```python
df.to_csv("diabetes_clean.csv",index=False)
```
